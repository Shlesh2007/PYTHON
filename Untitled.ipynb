```python
s1 = input('Enter first word : ')
s2 = input('Enter second word : ')

for i in s1:
    if i not in s2:
        print("Not Balanced")
        break
else:
    for i in s2:
        if i not in s1:
            print('Not Balanced')
            break
    else:
        print('Balanced')
        
```

```python
s = input('Enter a sentence : ')
s = s.title()[ :-1] + s[-1].upper()

for i,j in enumerate(s):
    if j == " ":
        count = i - 1
        break
s = s[0:count] + s[count].upper() + s[count+1:]
print(s)
```

```python
s = 'apple mango apple mango apple mango apple'
s.count('apple')              # 4
s.count('a')                  # 7
s.count('apple',1,len(s))     # 3
s.count('2')                  # 0

s.find('apple')              # 0
s.find('a')                  # 0
s.find('apple',1,len(s))     # 12
s.find('2')                  # -1

s.rfind('apple')             # 36
s.rfind('apple',1,36)        # 24
s.rfind('2')                 # -1
```
