```python
def reverse_num(n):
    "Reversing Number"
    summ=0
    while n>0:
        r=n%10
        summ=summ*10+r
        n=n//10
    return f"reverse Number: {summ}"

n=int(input("Enter number: "))
print(reverse_num(n))
```

```python
def cal(a,b):
    s=a+b
    d=a/b
    m=a*b
    return s,d,m#multiple return's ans always save in tuples
print(cal(3,5))
add,division,mul=cal(3,5)
print(add)
print(division)
print(mul)
```
