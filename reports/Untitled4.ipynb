```python
#Happy Number
n=int(input('Enter Number='))
while n!=1 and n!=4:
    summ=0
    while n>0:
        r=n%10
        summ=(r*r)+summ
        n=n//10
    n=summ
if n==1:
    print("It's Happy Number")
else:
    print("It's not a Happy Number")
```
