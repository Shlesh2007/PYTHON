```python
n=int(input("Enter Number="))
digit=len(str(n))
summ=0
temp=n
for i in range(digit):
        r=n%10
        summ=summ+(r**digit)
        n=n//10
if summ==temp:
    print(f"{temp} is Armstrong Number")
else:
    print(f"{temp} is not a Armstrong Number")
```

```python
n=int(input("Enter Number="))
digit=len(str(n))
summ=0
temp=n
for i in range(digit):
    r=n%10
    fact=1
    for j in range(1,r+1):
        fact=j*fact
    summ+=fact
    n=n//10
if summ==temp:
    print(f"{temp} is Strong Number")
else:
    print(f"{temp} is not a Strong Number")
```

```python
#learn jackong
max=0
min=0
c=0
summ=0
avg=0
while True:
  n=input("Enter Number=")
  if n=='Done':
        break
  else:
    c+=1
    summ+=int(n)
    if int(n)>max:
        max=int(n)
    elif int(n)<min:
        min=int(n)
if c==0:
    print("Pls Enter Numbers!!")
else: 
    avg=summ/c
    print(f"Minimum:{min} ")
    print(f"Maximum:{max}")
    print(f"Average:{avg}")
```
