```python
l=eval(input("Enter Ranking of Students in Tuple: "))
candy=[1]*len(l)
print(candy)
for i in range(1,len(l)):## l to r
    if l[i]>l[i-1]:#r to l
        candy[i]=candy[i-1]+1
for i in range(-2,-len(l)-1,-1):##r to l:
    if l[i]>l[i+1]:## l to r
        candy[i]=max(candy[i],candy[i+1]+1)
print(candy)  
print(sum(candy))
```
