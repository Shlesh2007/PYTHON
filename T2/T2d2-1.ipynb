```python
import string
##PASSWORD CHECK(identifier)
while True:
    uc=0
    lc=0
    dc=0
    sc=0
    spc=0
    p=input("Enter Password: ")
    if len(p)<8:
        print("The length of password should be minimum 8")
        continue
    else:
        flag=True
        for i,c in enumerate(p):
            if 'A'<=c and c<='Z':
                uc=uc+1
            else:
                lc=lc+1
            if c.isdigit():
                dc=dc+1
        if uc<2:
            print("The Password should have Atleast 2 capital letters")
            flag=False
            continue
        if lc<2:
            print("The Password should have Atleast 2 small letters")
            flag=False
            continue
        if dc<2:
            print("The Password should have Atleast 2 Digits")
            flag=False
            continue
        for i,c in enumerate(p): 
            if c<='A' or c>='Z' or c<='a' or c>='z':
                sc=sc+1
            if c.isspace():
                spc=spc+1
        if  sc<2:
                print("The Password should have Atleast 2 special characters")
                flag=False
                continue
        if spc>0:
                print("The Password should  not have spaces")
                flag=False
                continue
        if flag==True:
            print("Your Entered Password is Absoulatly Correct")
            break
        

            
        
```

```python
##ord(),chr()
msg=input("Enter message: ")
k=int(input("Enter Key Number: "))
s=""
count=0
for i,c in enumerate(msg):
        if c.isupper() or c.islower():
            l=ord(c)+k
        if c.isdigit():
            s=s+str(int(c)+k)
        if l>90 and l<97:
            l=l-90+64 
        if l>122:
            l=l-122+96
        if not c.isdigit():
            s=s+chr(l)
        if c.isspace():
            s=s+c

print(s)
```
