```python
##294
def cases(s):
    upper_case=0
    lower_case=0
    for i in s:
        if i.isupper():
            upper_case=upper_case+1
        elif i.islower():
            lower_case=lower_case+1
    return upper_case,lower_case
s=input("Enter Sring: ")
a,b=cases(s)
print(a,b)
```

```python
##310
def shifting(n,shift):
    n_s=""
    if shift>len(n):
        n_s=n[::-1]
    else:
        n_s=n[shift:]+n[:shift]
    return int(n_s)

n=int(input("Enter Number: "))
shift=int(input("Enter how much shifts you want? "))
n=int(shifting(str(n),shift))
print(n)
```

```python
##311
s=input("Enter String: ")
upper_case=lower_case=digits=0
for i in s:
    if i.isupper():
        upper_case+=1
    elif i.islower():
        lower_case+=1
    elif i.isdigit():
        digits+=1
print("No. of Uppercase: ",upper_case)
print("No. of Lowercase: ",lower_case)
print("No. of Digits: ",digits)
```

```python
#312
while True:
    s=input("Enter String: ")
    if len(s)<8:
        print("Length of password should be Minimum of 8 Characters")
        continue
    else:
        u_c=digit=spc=l_c=0
        for i in s:
            if i.isupper():
                u_c+=1
            elif i.islower():
                l_c+=1
            elif i.isdigit():
                digit+=1
            elif not(i.isalnum() or i.isspace()):
                spc+=1
        print(u_c,digit,spc)
        if (u_c>=1 and digit>=1 and spc>=1) and (u_c+digit+spc+l_c==len(s)):
            print("Coorect Password!!")
            break
        else:
            print("Incorrect Password,Try Again!!")
```

```python
s=input("Enter String: ")
n=int(input("Enter Shifts: "))
n_s=""
for i in s:
    if i.islower():
        l=ord(i)
        if l>=122:
            n_s+=chr(l-122+96+n)
        else:
            n_s+=chr(l+n)
    elif i.isupper():
        l1=ord(i)
        if l1>=90:
            n_s+=chr(l1-90+64+n)
        else:
            n_s+=chr(l1+n)
    else:
        n_s+=i
print(n_s)    
```
