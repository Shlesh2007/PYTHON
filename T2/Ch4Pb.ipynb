```python
##454
stack = []
d = {'}':'{' , ']':'[' , ')':'('}

def f(bracket):
    for i in bracket:
        if i in '[({':
            stack.append(i)
        else:
            if len(stack)==0:
                return False
            elif d[i]!=stack.pop():
                return False
    if len(stack)!=0:
        return False
    else:
        return True
bracket =input("Enter Brackets: ")
print('Brackets are Balanced' if f(bracket) else 'Brackets are not balanced')
```
