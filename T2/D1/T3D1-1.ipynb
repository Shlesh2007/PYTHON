```python
f=open('sample.txt','w+')##w,w+ makes file clear(write+read)
print(f.tell())## to now curren position of pointer
f.write('welcome')
print(f.tell())
##to set pointer seek(offset,position) positions:- 0-start,1-current,2-end supports only in binary file but (0,1) and (0,2) are supported
f.seek(3,0)
print(f.tell())
print(f.read())
print(f.tell())
f.close()
```

```python
f=open('sample.txt','a+')##w,w+ makes file clear(write+read)
print(f.tell())## to now curren position of pointer
f.write('welcome')
print(f.tell())
f.seek(0)
print(f.read())
print(f.tell())
f.close()
```

```python
##not in syllabus
with open('sample.txt') as f:##automatic open and close
    ##inside indent block the file is opened
    for i in f:
        print(i)
f.read()
##outside indent block the file is closed
```

```python
##not in syllabus
with open('sample.txt','r') as f:##automatic open and close
    ##inside indent block the file is opened
    print(f.name)
    print(f.mode)
    print(f.closed)
    print(f.readable())
    print(f.writable())
##outside indent block the file is closed
```
