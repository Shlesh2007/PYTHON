```python
L = eval(input("Enter array : "))

left = [0] * len(L)
right = [0] * len(L)

left[0] = L[0]

for i in range(1,len(L)):
    left[i] = max(L[i],left[i-1])
    
right[-1] = L[-1]

for j in range(len(L) - 2,-1,-1):
    right[j] = max(L[j],right[j+1])
    
ans = 0

for i in range(len(L)):
    ans = ans + min(left[i],right[i])-L[i]
    
print(ans)
    
```
