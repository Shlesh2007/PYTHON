```python
s=eval(input("Enter Array(Water trapping): "))
left=[0]*len(s)
right=[0]*len(s)
left[0]=s[0]
for i in range(1,len(s)):
    s[i]=max(s[i],left[i-1])
right[-1]=s[-1]
for j in range(len(s)-2,-1,-1):
    s[j]=max(s[j],right[j+1])
ans=0
for i in range(len(s)):
    ans+=min(left[i],right[i])-s[i]
print(ans)
```
