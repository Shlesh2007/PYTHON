```python
class A:
    def __init__(self):
        print("Hello")
    def show(self):
        print("Bye")
class B(A):
    def __init__(self):
        print("Shlesh")
        super().__init__()
    def show(self):
        print("Python")
        super().show()

b=B()
b.show()
```

```python
class A:
    def __init__(self,a,b,c):
        print(a,b,c)
    def show(self):
        print("Bye")
class B(A):
    def __init__(self,d,a,b,c):
        self.d=d
        print(self.d,end=" ")
        super().__init__(a,b,c)
    def show(self):
        print("Python")
b=B(1,2,3,4)
b.show()
```
