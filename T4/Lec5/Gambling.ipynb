```python
from numpy import random
game_times=int(input("How many Times Do you want to play this game? "))
game_balance=1000

game_count=1
while game_count<=game_times:
    if game_balance>0:
        n=int(input("Enter Number= "))
        randome_no=random.randint(n-1)
        if randome_no==3:
            if n==3:
                print("You won the Game!!")
                game_balance+=100
            else:
                print("You Lost The Game,Better luck next time")
                game_balance-=100
        elif randome_no<3:
            if n<3:
                print("You won the Game!!")
                game_balance+=100
            else:
                print("You Lost The Game,Better luck next time")
                game_balance-=100
        if randome_no>3:
            if n>3:
                print("You won the Game!!")
                game_balance+=100
            else:
                print("You Lost The Game,Better luck next time")
                game_balance-=100
        game_count+=1
    else:
        print("Sorry you can't play the next game sue to Insufficient Balance")
        break
if game_balance>0:
    print("Final Balance",game_balance)
```
