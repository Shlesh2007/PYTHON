```python
from numpy import random
import numpy as np
scores=random.randint(1,100,size=(5,4))
print(scores)
```

```python
bats1=random.randint(1,100,size=(1,4))
bats2=random.randint(1,100,size=(1,4))
print(bats1)
print(bats2)
```

```python
#concatenate axis=0=>Row,axis=1=>Column
#min,max axis=1=>Row,axis=0=>Column
b1=np.concatenate((scores,bats1),axis=0)
b2=np.concatenate((b1,bats2),axis=0)

print(b2)
```

```python
T20_5=random.randint(1,100,size=(7,1))
final_match=np.concatenate((b2,T20_5),axis=1)
print(final_match)
```

```python
summ=np.sum(final_match,axis=1)
s=summ.reshape(7,1)
# print(s)
f_match=np.concatenate((final_match,s),axis=1)
print(f_match)
```

```python
maxi=np.max(final_match[:,:-1],axis=1)
print(maxi)
tests=["T20_1","T20_2","T20_3","T20_4","T20_5"]
```
