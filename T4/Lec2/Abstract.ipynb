```python
from abc import ABC,abstractmethod##ABC abstract class is by default made in python
class defence(ABC):
    @abstractmethod
    def gun(self):
        pass
    def fire(self):
        print("Firing")
class Army(defence):
    def gun(self):
        print("Gun is obtained")
    def defen(self):
        print("Fj")
a=Army()
a.gun()
a.fire()
a.defen()
```

```python
from abc import ABC,abstractmethod
class Employee(ABC):
    def getdata(self):
        self.emp_id=int(input("Enter Employee ID:"))
        self.emp_name=input("Enter Employee name:")
        self.emp_salary=float(input("Enter Employee salary:"))
        
    def empdata(self):
        print("Employee ID:",self.emp_id)
        print("Employee name:",self.emp_name)
        print("Employee salary:",self.emp_salary)
    @abstractmethod
    def emp_id(self):
        pass
    def empsalary(self):
        return self.emp_salary
class Perks(Employee):
    def calculate(self):
        self.da=self.emp_salary*0.35
        self.hra=self.emp_salary*0.17
        self.pf=self.emp_salary*0.12
    def getcalculated(self):
        print("DA:",self.da)
        print("HRF:",self.hra)
        print("PF:",self.pf)
        self.total_perks=self.da+self.hra-self.pf
class NetSalary(Perks):
    def totalSalary(self):
        self.total_salary=self.total_perks+self.emp_salary
        self.empdata()
        print("Total Salary",self.total_salary)
        
    def emp_id(self):
        pass
m=NetSalary()
m.getdata()
m.calculate()
m.getcalculated()
m.totalSalary()
```
