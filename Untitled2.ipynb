```python
# penny=1,fives=5,tens=10,quarter=25,nickle=75
n=int(input("Enter Number:"))
if n>=75:
    nic=n//75
    n=n%75
    print(f"{nic} nickle",end=" ")
if n>=25:
    nic=n//25
    n=n%25
    print(f"{nic} quarter",end=" ")
if n>=10:
    nic=n//10
    n=n%10
    print(f"{nic} tens",end=" ")
if n>=5:
    nic=n//5
    n=n%5
    print(f"{nic} fives",end=" ")
if n>=1:
    nic=n//1
    n=n%1
    print(f"{nic} penny",end=" ")

```
