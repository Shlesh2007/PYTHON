```python
# 529 Write a Python class named Circle constructed by a radius and 
# two methods which will compute the area and the perimeter of a circle.
class circle:
    def __init__(self,r):
        self.p=3.14
        self.r=r
        self.area()
        self.perimeter()
    def area(self):
        a=self.p*self.r*self.r
        print("Area of the circle is =",a)
    def perimeter(self):
        p=2*self.p*self.r
        print("Perimeter of the circle is =",p)
    def __del__(self):
        print("Bye")
c=circle(4)
del c
```

```python
# create_pin,change_pin,menu,withdrawal(min 1000 should remain),exit
class Atm():
    def __inint__(self):
        self.pin=0
        self.balance=0
        self.old_pin=0
    def create_Pin(self):
        self.pin=int(input("Create New Pin: "))
        self.balance=int(input("Enter Balance: "))
        print("Pin Setted Successfully!!\n")
    def change_pin(self):
        self.old_pin=int(input("Enter old Pin: "))
        if self.pin==self.old_pin:
            self.pin=int(input("Reset New Pin: "))
        else:
            print("Wrong Pin Entered!!\n")
    def withdrawal(self,bal):
        if (self.balance-bal<1000):
            print("Can't Withdraw since remaining balance will be less than 1000\n")
        elif bal>self.balance:
            print("Withdrawal Money is Greater than the Account Balance\n")
        else:
            self.balance=self.balance-bal
            print("Money has been withdrawed\n")
    def menu(self):
        while True:
            print("1.Create Pin")
            print("2.Set Pin")
            print("3. Withdraw Money")
            print("4. Exit\n")
            ch=int(input("Enter Choice:"))
            if ch==4:
                print("Exiting...")
                return
            else:
                if ch==1:
                    self.create_Pin()
                elif ch==2:
                    self.change_pin()
                elif ch==3:
                    w_money=int(input("Enter Withdrawal Money: "))
                    self.withdrawal(w_money)
Acc=Atm()
Acc.menu()
```
