```python
import math
math.factorial(7)
n=int(input("Enter R="))
for i in range(n):
    for j in range (n+1,i,-1):
        print(" ",end=" ")
    for k in range(0,i+1):
        print(" ",math.factorial(i)//(math.factorial(k)*math.factorial(i-k)),end=" ")
    print()
```
